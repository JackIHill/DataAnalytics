```python
import pandas as pd
import os
```

```python
cwd = os.getcwd()
```

```python
data_original = pd.read_csv('data_cleaning_challenge.csv')
data_original.drop(columns=['Unnamed: 9', 'Unnamed: 10'], inplace=True)
```

```python
dropped_nans = data_original[data_original['Row Type'].notna()]
dropped_nans
```

```python
col_values = []
counter = 0

for i in dropped_nans['Row Type']:
    if 'first name' in i:
        counter += 1
    col_values.append(counter)
```

```python
data_iter = dropped_nans.copy()
data_iter['Iteration'] = col_values
data_iter
```

```python
drop_extra = data_iter[data_iter['Row Type'] != 'Row Type']
```

```python
name_data = drop_extra[drop_extra['Row Type'].str.contains('first name')]
name_data
```

```python
name_data_clean = name_data.drop(columns=['Speed1', 'Speed2', 'Electricity', 'Effort', 'Weight', 'Torque'])
name_data_clean.rename(columns={'Row Type': 'First Name', 'Iter Number': 'Last Name', 'Power1': 'Date'}, inplace=True)
name_data_clean['First Name'] = name_data_clean['First Name'].str[12:]
name_data_clean['Last Name'] = name_data_clean['Last Name'].str[11:]
name_data_clean['Date'] = name_data_clean['Date'].str[6:]
name_data_clean
```

```python
no_name_data = drop_extra[~drop_extra['Row Type'].str.contains('first name')] # ~ = opposite
no_name_data
```

```python
pd.merge(left=name_data_clean, right=no_name_data, how='inner', on='Iteration')
```
