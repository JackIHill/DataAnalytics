```python
import pandas as pd
import os
```

```python
sales_data = [f for f in os.listdir('./Sales_Data')]

all_months = pd.DataFrame()

for data in sales_data:
    df = pd.read_csv('./Sales_Data/' + data)
    all_months = pd.concat([all_months, df])
    
all_months.to_csv('all_months.csv', index=False)
```

Read in updated dataframe (so don't have to reload cell above multiple times)

```python
data = pd.read_csv('all_months.csv')
```

### Clean up NaN's

```python
nan_df = data[data.isna().any(axis=1)]
```

```python
data_nonan = data.dropna(how='all').copy()
data_nonan['Month'] = data_nonan['Order Date'].str[:2]
```

### Question: Which month performed the best? What was the total earnings for that month?

```python
all_data = data_nonan[data_nonan['Month'] != 'Or'].copy()

all_data['Month'] = all_data['Month'].astype('int32')
all_data['Quantity Ordered'] = all_data['Quantity Ordered'].astype('int32')
all_data['Price Each'] = all_data['Price Each'].astype('float')
all_data
```

#### Add Sales Column

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data
```

```python
month_group = all_data.groupby('Month').sum()
best_month = month_group.loc[month_group['Sales'].idxmax()]
best_month.reset_index()
best_month
```

```python
month_group
```

```python
import matplotlib.pyplot as plt
# can use this as x axis also.
months = list(range(1, 13))

plt.plot(months, month_group['Sales'], color='k')

plt.xticks(months)
plt.xlabel('Month Number')
plt.ylabel('Sales in USD ($)')
plt.title('Sales (USD) per Month')
plt.ticklabel_format(style='plain') # prevents scientific notation (prints full sales value)

plt.show()
```

#### Which city had the highest number of sales?

```python
all_data['City'] = all_data['Purchase Address'].apply(
    lambda x: f"{x.split(',')[1]},{x.split(',')[2][0:3]}")

state = all_data

city_group = all_data.groupby('City').sum()
cities = city_group.index.to_numpy()
```

```python
plt.bar(cities, city_group['Sales'], color='k')

plt.xlabel('City')
plt.ylabel('Sales (USD)')

plt.xticks(rotation=45, ha='right') # diagonal xticks so don't block each other
plt.ticklabel_format(style='plain', axis='y') # have to define axis as can't format strings that way)

plt.title('Sales (USD) per City')
plt.show()
```

### What time should we display advertisements to maximise likelihood of customers buying product?

```python
time_data = all_data.copy()
time_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
time_data['Hour'] = time_data['Order Date'].dt.hour
time_data['Minute'] = time_data['Order Date'].dt.minute
time_data.drop(columns=['Order Date', 'City', 'Purchase Address'], inplace=True)
```

```python
best_time = time_data.groupby(time_data['Hour']).count()

plt.plot(best_time.index.to_numpy(), best_time['Quantity Ordered'])

plt.xticks(best_time.index.to_numpy())
plt.xlabel('Hour')

plt.ylabel('Quantity Ordered')
plt.title('Quantitiy of Product Ordered per Hour')

plt.grid()
plt.show()
```

### What products are most often sold together?

```python
all_data.head()
```

```python
# get df where order ID has occured more than once (i.e two or more products were sold)
order_together = all_data[all_data['Order ID'].duplicated(keep=False)].copy()

order_together['Grouped'] = order_together.groupby(
    'Order ID')['Product'].transform(lambda x: ', '.join(x))

order_together = order_together[['Order ID', 'Grouped']].drop_duplicates()
order_together
```

```python
from itertools import combinations
from collections import Counter

count = Counter()
for row in order_together['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

count.most_common(10)
```

### What product sold the most? Why do you think so?

```python
best_product = all_data.groupby('Product').sum().reset_index()
best_product.sort_values('Quantity Ordered', ascending=False)
```

```python
plt.bar(best_product['Product'], best_product['Quantity Ordered'])

plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
plt.xticks(rotation='vertical')
plt.title('Quantity Ordered by Product')

plt.show()
```

```python
prices = all_data.groupby('Product')['Price Each'].mean().reset_index()

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(best_product['Product'], best_product['Quantity Ordered'])
ax2.plot(prices['Product'], prices['Price Each'], color='k')

ax1.set_xlabel('Product')
ax1.set_ylabel('Quantity Ordered', color='b')
ax2.set_ylabel('Price Each (USD/$)')
ax1.set_xticklabels(labels=best_product['Product'], rotation='vertical')
plt.title('Quantity Ordered by Product')

plt.show()
```
