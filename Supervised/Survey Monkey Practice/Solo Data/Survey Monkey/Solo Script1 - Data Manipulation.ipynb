```python
import pandas as pd
import os
```

```python
cwd = os.getcwd()
```

```python
data = pd.read_excel(cwd + '\\Solo Data\\Solo Data - Raw Survey Monkey Output.xlsx', sheet_name='Edited_data')
```

```python
data = data.drop(columns=['Start Date', 'End Date', 'Email Address', 'First Name', 'Last Name', 'Custom Data 1 '])
```

```python
id_vars = list(data.columns[:8])
value_vars = list(data.columns[8:])

data_melted = pd.melt(data, id_vars=id_vars, value_vars=value_vars, var_name='Question+Subquestion', value_name='Answer')
data_melted
```

```python
question_data = pd.read_excel('Solo Data\\Solo Data - Raw Survey Monkey Output.xlsx', sheet_name='Questions')
question_data.drop(columns=['Response', 'Question'], inplace=True)
question_data.rename(columns={'Question + Subquestion': 'Question+Subquestion'}, inplace=True)
```

```python
data_merged = data_melted.copy()
data_merged = pd.merge(left=data_merged, right=question_data, left_on='Question+Subquestion', right_on='Question+Subquestion')
data_merged.rename(columns={'Question.1': 'Question', 'Respondent ID ': 'RespondentID'}, inplace=True)
data_merged
```

```python
respondents = data_merged[data_merged['Answer'].notna()]
respondents = respondents.groupby('Question')['RespondentID'].nunique().reset_index()
respondents.rename(columns={'RespondentID': 'Total Respondents'})
```

```python
data_merged_2 = data_merged.copy()
data_merged_2 = pd.merge(left=data_merged_2, right=respondents, left_on='Question', right_on='Question')
```

```python
data_merged_2.rename(columns={'RespondentID_x': 'RespondentID', 'RespondentID_y': 'Total Respondents'}, inplace=True)
```

```python
same_ans = data_merged_2.copy()
same_ans = same_ans.groupby(['Question+Subquestion', 'Answer'])['RespondentID'].nunique().reset_index()
same_ans.rename(columns={'RespondentID': 'Same Answer'}, inplace=True)
```

```python
data_merged_3 = data_merged_2.copy()
data_merged_3 = pd.merge(left=data_merged_2, right=same_ans, left_on=['Question+Subquestion', 'Answer'], right_on=['Question+Subquestion', 'Answer'])
data_merged_3
```

```python
final_data = data_merged_3.copy()
final_data.rename(columns={
    'Identify which division you work in. - Response': 'Primary Division',
    'Identify which division you work in. - Other (please specify)': 'Secondary Division',
    'Which of the following best describes your position level? - Response': 'Position',
    'Which generation are you apart of? - Response': 'Generation',
    'Please select the gender in which you identify. - Response': 'Gender',
    'Which duration range best aligns with your tenure at your company? - Response': 'Company Tenure',
    'Which of the following best describes your employment type? - Response': 'Employment Type',
    'Question+Subquestion': 'Question and Subquestion'
    }, inplace=True)
```

```python
final_data.to_excel(cwd + '\\Final Transformed Data.xlsx', index=False)
```
